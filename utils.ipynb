# Visualizing .obj

```python
import numpy as np 

def load_obj(file_path):
    "Get Vertices and triangles from a .obj file"
    vertices = []
    faces = []

    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith('v '):  # vertex
                parts = line.strip().split()
                v = list(map(float, parts[1:4]))
                vertices.append(v)

            elif line.startswith('f '):  # face
                parts = line.strip().split()
                
                # f 1 2 3  veya  f 1/1/1 2/2/2 3/3/3
                face = []
                for p in parts[1:]:
                    idx = p.split('/')[0]  # sadece vertex index al
                    face.append(int(idx) - 1)  # obj 1-based → 0-based

                faces.append(face)

    return np.array(vertices), np.array(faces)
```

```python
file_path = "source_obj/cat-poses/cat-01.obj"
v,f = load_obj(file_path=file_path)
v.shape, f.shape
```
